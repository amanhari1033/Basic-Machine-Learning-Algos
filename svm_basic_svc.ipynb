```python
from sklearn.svm import SVC
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split

x,y = load_iris(return_X_y = True)

# constraining to two classes for better visualization
x = x[1:900,0:2]
y = y[1:900]

color1 = (1,0,0,1)
color2 = (0,0,1,1)
color3 = (0,1,0,1)

colormap = np.array([color1,color2,color3])
plt.scatter(x[:,0],x[:,1],c=colormap[y])
plt.grid(True)
```

```python
#using linear kernel
clf = SVC(kernel ='linear')
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
clf.fit(xtrain,ytrain)
clf.score(xtest,ytest)
p = np.linspace(min(x[:,0]),max(x[:,0]),200)
q = np.linspace(min(x[:,1]),max(x[:,1]),200)

m,n = np.meshgrid(p,q)
z = np.zeros((200,200))
for i in range(np.shape(m)[0]):
    for j in range(np.shape(n)[0]):
        z[i,j] =  clf.predict(np.reshape([m[i,j],n[i,j]],(1,2)))
plt.pcolor(m,n,z,cmap ='Greys')
plt.scatter(x[:,0],x[:,1],c=colormap[y])
plt.grid(True)
plt.show()
```

```python
#using kernel poly
clf = SVC(kernel ='poly',C=10)
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
clf.fit(xtrain,ytrain)
print('Accuracy using poly kernel:'+str(clf.score(xtest,ytest)))

z = np.zeros((200,200))
for i in range(np.shape(m)[0]):
    for j in range(np.shape(n)[0]):
        z[i,j] =  clf.predict(np.reshape([m[i,j],n[i,j]],(1,2)))
        
plt.pcolor(m,n,z,cmap ='Greys')
plt.scatter(x[:,0],x[:,1],c=colormap[y])
plt.grid(True)
plt.show()
```

```python
# using rbf kernel
# vary gamma parameter to avoid overfitting
from sklearn import svm
clf = svm.NuSVC(gamma='auto')
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
clf.fit(xtrain,ytrain)
print('Accuracy using poly kernel:'+str(clf.score(xtest,ytest)))

z = np.zeros((200,200))
for i in range(np.shape(m)[0]):
    for j in range(np.shape(n)[0]):
        z[i,j] =  clf.predict(np.reshape([m[i,j],n[i,j]],(1,2)))
        
plt.pcolor(m,n,z,cmap ='Greys')
plt.scatter(x[:,0],x[:,1],c=colormap[y])
plt.grid(True)
plt.show()
```
