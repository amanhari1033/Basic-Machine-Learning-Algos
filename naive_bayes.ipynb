```python
from sklearn.datasets import load_breast_cancer
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

x,y = load_breast_cancer(return_X_y=True)

# constraining x to only 2 features for visualization
# schowing only 100 data for ease of visualization
x = x[0:200,0:2]
y = y[0:200]
plt.scatter(x[:,0],x[:,1],c=y)
plt.grid(True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
clf = GaussianNB()
clf.fit(x_train,y_train)
print("Accuracy:"+str(clf.score(x_test,y_test)))
```

```python
import numpy as np
p = np.linspace(min(x[:,0]),max(x[:,0]),100)
q = np.linspace(min(x[:,1]),max(x[:,1]),100)

m,n = np.meshgrid(p,q)
z = np.zeros((100,100))
for i in range(np.shape(m)[0]):
    for j in range(np.shape(n)[0]):
        z[i,j] =  clf.predict(np.reshape([m[i,j],n[i,j]],(1,2)))
```

```python
plt.pcolor(m,n,z,cmap='Set1')
plt.scatter(x[:,0],x[:,1],c = y+10)
plt.grid(True)
plt.show()
```
